```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

```python
audio_path = 'Recording.wav'
y, sr = librosa.load(audio_path)
audio_path2 = 'Recording1.wav'
y2, sr2 = librosa.load(audio_path2)
audio_path3 = 'Recording2.wav'
y3, sr3 = librosa.load('Recording2.wav')
```

```python
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
mfccs2 = librosa.feature.mfcc(y=y2, sr=sr2, n_mfcc=13)
mfccs3 = librosa.feature.mfcc(y=y3, sr=sr3, n_mfcc=13)
```

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, x_axis='time', cmap='viridis')
plt.colorbar(format='%+2.0f dB')
plt.title('MFCC')
plt.xlabel('Time')
plt.ylabel('MFCC Coefficient')
plt.show()

plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs2, x_axis='time', cmap='viridis')
plt.colorbar(format='%+2.0f dB')
plt.title('MFCC 2')
plt.xlabel('Time')
plt.ylabel('MFCC Coefficient')
plt.show()

plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs3, x_axis='time', cmap='viridis')
plt.colorbar(format='%+2.0f dB')
plt.title('MFCC 3')
plt.xlabel('Time')
plt.ylabel('MFCC Coefficient')
plt.show()
```
